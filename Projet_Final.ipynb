**Dataset Information**

There are seven variables in our current data set. Among these variables, species, island and sex are categorical, and the remaining variables are numeric variables. The dataset contains information about penguins. This information in order;

> **Species:** Types of penguins in the dataset

> **Island:** Islands of penguins in the dataset

> **Culmen_length_mm:** Culmen(bill) lengths in millimeters of penguins in the data set

> **Culmen_depth_mm:** Culmen(bill) depth in millimeters of penguins in the data set

> **Flipper_length:** Flipper lengths in millimeters of penguins in the data set

> **Body_mass_g:** Weights in grams of penguins in the data set

> **Sex:** The genders of the penguins in the data set

```python
#Connaitre le pourcentage de NaN selon le nombre de cellules du DataFrame
nb_cells = np.product(df.shape)
nb_na = df.isna().sum().sum()
nb_na / nb_cells * 100
```

```python
def count_na(df) : 
    mask = df.isna().sum() > 0
    missing_values_count = df.isna().sum()
    return missing_values_count[mask].sort_values(ascending=False)
```

```python
def count_na_percent(df, threshold) : 
    mask = df.isna().sum() > threshold
    missing_values_count = df.isna().sum()
    return (missing_values_count[mask] / len(df) *100).sort_values(ascending=False)
```

```python
count_na_percent(df, 0)
```

```python
missing_values_count = count_na_percent(df[categorical_cols], 0)
missing_values_count.sort_values(ascending=False)
```

```python
missing_values_count = count_na_percent(df[numerical_cols], 0)
missing_values_count.sort_values(ascending=False)
```

```python
df = df.drop(["Comments"], axis=1)
```

```python
# Add Culmen Surface (mm2), Total Lenght (mm), Delta Tot (o/oo)
df["Culmen Surface (mm2)"] = df["Culmen Length (mm)"] * df["Culmen Depth (mm)"]
df["Total Lenght (mm)"] = df["Culmen Length (mm)"] + df["Culmen Depth (mm)"] + df["Flipper Length (mm)"]
df["Delta Tot (o/oo)"] = df["Delta 15 N (o/oo)"] + df["Delta 13 C (o/oo)"]
```

```python
df_sex = df
# df_sex.dropna()
df_sex["Sex"] = df_sex["Sex"].replace(np.nan, 'None Specified')
df_sex["Sex"] = df_sex["Sex"].replace('.', 'None Specified')
```

```python
df_na=df.dropna()
```

```python
X=pd.get_dummies(df_na.drop('Species',axis=1))
y=df_na.Species
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix

print(classification_report(y_test,base_pred))
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=200)
plot_tree(model,feature_names=X.columns,filled=True)
```

```python
def report_model(model):
    model_pred=model.predict(X_test)
    print(classification_report(y_test,model_pred))
    print('\n')
    plt.figure(figsize=(12,8),dpi=200)
    plot_tree(model,feature_names=X.columns,filled=True)
```

```python
report_model(model)
```

```python
pruned_tree=DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```
